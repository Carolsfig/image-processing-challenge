```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
import numpy as np

# Passo 1: Fazer upload da imagem
uploaded = files.upload()

# Pega o nome do arquivo enviado
filename = list(uploaded.keys())[0]

# Passo 2: Ler a imagem
img_colorida = cv2.imread(filename)

# Passo 3: Converter para cinza
img_cinza = cv2.cvtColor(img_colorida, cv2.COLOR_BGR2GRAY)

# Passo 4: Binarizao (preto e branco)
_, img_binaria = cv2.threshold(img_cinza, 127, 255, cv2.THRESH_BINARY)

# Passo 5: Exibir resultados
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Colorida")
plt.imshow(cv2.cvtColor(img_colorida, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Cinza")
plt.imshow(img_cinza, cmap="gray")
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Binria")
plt.imshow(img_binaria, cmap="gray")
plt.axis("off")

plt.show()
```
